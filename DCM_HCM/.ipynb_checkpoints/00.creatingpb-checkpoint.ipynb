```python
#!/usr/bin/env python
# coding: utf-8
import sys
import scanpy as sc
import anndata
import pandas as pd
import numpy as np
import os
import gc
import decoupler as dc
import regex as re
os.chdir("/Users/ricardoramirez/Dropbox/PhD/Research/MOFAcell")
```

```python
sc_dat = sc.read_h5ad(filename = "./data_HCMDCM/counts.h5ad")
```

```python
available_cells = sc_dat.obs[["cell_type_leiden0.6"]].drop_duplicates()
```

```python
available_cells
```

```python
available_cells["cell_type"] = "none"
```

```python
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Cardiomyocyte'), 'cell_type'] = "CM"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Adipocyte'), 'cell_type'] = "Adipo"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Macrophage', flags=re.IGNORECASE), 'cell_type'] = "Myeloid"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Fibroblast', flags=re.IGNORECASE), 'cell_type'] = "Fib"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('VSMC'), 'cell_type'] = "vSMCs"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Endothelial', flags=re.IGNORECASE), 'cell_type'] = "Endo"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Mast', flags=re.IGNORECASE), 'cell_type'] = "Mast"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Pericyte', flags=re.IGNORECASE), 'cell_type'] = "PC"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Lymphocyte', flags=re.IGNORECASE), 'cell_type'] = "Lymphoid"
available_cells.loc[available_cells["cell_type_leiden0.6"].str.contains('Neuronal', flags=re.IGNORECASE), 'cell_type'] = "Neuronal"
```

```python
available_cells
```

```python
sc_dat.obs["barcode"] = sc_dat.obs.index.values
```

```python
new_cts = sc_dat.obs.merge(available_cells, on='cell_type_leiden0.6', how='left')[["barcode","cell_type"]]
```

```python
sc_dat.obs.loc[new_cts["barcode"].values, "cell_type"] = new_cts["cell_type"].values
```

```python
sc_dat.obs
```

```python
meta_data = sc_dat.obs[["donor_id", "disease","sex", "age", "lvef"]].drop_duplicates()
```

```python
meta_data.to_csv("./data_HCMDCM/meta_data.csv")
```

```python
cell_type_numbers = sc_dat.obs.groupby(["donor_id", "cell_type"])["cell_type"].count()
cell_type_numbers = cell_type_numbers.to_frame().rename({"cell_type":"counts"}, axis=1).reset_index()
```

```python
sc_dat.layers['counts'] = sc_dat.X
```

```python
padata = dc.get_pseudobulk(sc_dat, sample_col='donor_id', groups_col='cell_type', layer='counts', min_prop=0, min_smpls=0)
```

```python
padata.var.index.values
```

```python
pb_dat = pd.DataFrame(padata.X)
pb_dat.columns = padata.var.index.values
pb_dat.index = padata.obs.index.values
pb_dat.to_csv("./data_HCMDCM/pb_data.csv")
```

```python
pb_coldata = padata.obs
pb_coldata["colname"] = pb_coldata.index.values
pb_coldata = pb_coldata.merge(cell_type_numbers, on = ["cell_type","donor_id"], how = "left")
```

```python
pb_coldata.to_csv("./data_HCMDCM/pb_coldata.csv")
```
